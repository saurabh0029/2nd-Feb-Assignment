```python
Q1. Ans

When do We use for loops

Actually, for loops are used when we have a block of code which we want to repeat a fixed number of times 
and for-loop is always used in combination with an iterable object, like a list or a range. The Python 
for statement iterates over the members of a sequence in order, executing the block each time. 

For example , Suppose We have to print tables then we already knew that the no of iteration will be 10,
Here we will use for loops.

n=int(input("Enter a Number"))
i=range(1,11)
result=0
for i in i :
    result=n*i
    print(n, "X",i, "=",result)


    
When to use while loops

Actually, the while loop is used when it is impossible to determine the exact number of loop iterations in advance.
for example ,
Check how many times a given number can be divided by 3 before it is less than or equal to 10

In this example we don't have any idea , how many time will the loop execute.
In this case we should use while loop 

count=0
n=int(input("Enter a Number"))

while n>=10:
    count=count+1
    n=n/3
print(count,"times can be divided by 3 before it is less than or equal to 10.")
```

```python
Q2. Ans

# Using For Loop

su_m=0
product=1

i=range(1,11)

for k in i :
    su_m=su_m+k
    product=product*k

print("Sum of first 10 natural numbers is",su_m,"\nProduct of first 10 natural number is",product)    

# Using while loop

su_m=0
product=1
i=1

while i<=10:
    su_m=su_m+i
    product=product*i
    i=i+1

print("Sum of first 10 natural numbers is",su_m,"\nProduct of first 10 natural number is",product)    
```

```python
Q4. Ans

# Using While and For loop

i=1

l_cube=[]
l_divisible_by_4_or_5=[]

while i <=100:
    l_cube.append(i**3)
    i=i+1
else:
    for j in l_cube:
        if j%4==0 or j%5==0:
            l_divisible_by_4_or_5.append(j)
print("Cube of Each numbers\n\n",l_cube)
print("\n\n Divisible by 4 or 5\n\n",l_divisible_by_4_or_5)

# Using only For Loop


l=list(range(1,101))

l_cube=[]
l_divisible_by_4_or_5=[]

for i in l:
    l_cube.append(i**3)
else:
    for j in l_cube:
        if j%4==0 or j%5==0:
            l_divisible_by_4_or_5.append(j)
print("Cube of Each numbers\n\n",l_cube)
print("\n\n Divisible by 4 or 5\n\n",l_divisible_by_4_or_5)
    
```
